```python
pcpSummary = pd.read_csv("./CMADS/For-swat-2012/Fork/PCPFORK.txt")
```

```python
pcpSummary.head()
```

```python
max_longitude = 135
min_longitude = 73
max_latitude = 54
min_latitude = 18

pcpSummary = pcpSummary[(pcpSummary['LONG'] <= max_longitude) & (pcpSummary['LONG'] >= min_longitude) & (pcpSummary['LAT'] <= max_latitude) & (pcpSummary['LAT'] >= min_latitude)]
pcpSummary.describe()
```

```python
from multiprocessing import Pool


def read_pcp_data(path, fileList):
    pcp_data = pd.DataFrame()
    # each file a column, different file in different columns, 
    columns = []
    for file in fileList:
        
        # pcp_data = pd.concat([pcp_data, pd.read_csv(os.path.join(path, file)+".txt", header=None)[1:]], axis=1)
        
        temp = pd.read_csv(os.path.join(path, file)+".txt", header=None)[1:]
        columns.append(temp)
        
    pcp_data = pd.concat(columns, axis=1)
    pcp_data.columns = fileList
    date = pd.date_range(start='2008-01-01', end='2016-12-31', freq='D')
    pcp_data.set_index(date, inplace=True)
    return pcp_data
    
    


path = "./CMADS/For-swat-2012/Station/Precipitation-104000-txt"
fileList = pcpSummary.NAME



pcp_data = read_pcp_data(path, fileList)
```

```python
pathTemp = "./CMADS/For-swat-2012/Station/Temperature-104000-txt"
tempSummary = pd.read_csv("./CMADS/For-swat-2012/Fork/TMPFORK.txt")

tempSummary = tempSummary[(tempSummary['LONG'] <= max_longitude) & (tempSummary['LONG'] >= min_longitude) & (tempSummary['LAT'] <= max_latitude) & (tempSummary['LAT'] >= min_latitude)]
tempSummary.head()
```

```python
def read_tmp_data(path, fileList):
    pcp_data = pd.DataFrame()
    # each file a column, different file in different columns, 
    columns = []
    for file in fileList:
        
        # pcp_data = pd.concat([pcp_data, pd.read_csv(os.path.join(path, file)+".txt", header=None)[1:]], axis=1)
        # print(os.path.join(path, file))
        temp = pd.read_csv(os.path.join(path, file)+".txt", sep=",",skiprows=1, header=None)
        temp = temp.mean(axis=1)
        columns.append(temp)
        
    pcp_data = pd.concat(columns, axis=1)
    pcp_data.columns = fileList
    date = pd.date_range(start='2008-01-01', end='2016-12-31', freq='D')
    pcp_data.set_index(date, inplace=True)
    return pcp_data


fileList = tempSummary.NAME

temp_data = read_tmp_data(pathTemp, fileList)

temp_data.info()
```
